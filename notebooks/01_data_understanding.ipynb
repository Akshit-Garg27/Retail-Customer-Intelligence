# 1. Business Context

A retail company has accumulated customer demographic, purchasing, and campaign response data over several years. While this information is available, marketing decisions are currently broad and largely untargeted, resulting in inefficient promotional spending and inconsistent customer engagement.

The company wants to better understand its customers by identifying meaningful customer groups based on purchasing behavior, demographics, and campaign interactions. These insights will support more personalized marketing strategies, stronger customer retention, and improved allocation of marketing resources.

Rather than treating every customer the same, the objective is to transform historical customer data into actionable business intelligence that supports strategic decision-making.

# 2. Problem Satetment
The current marketing strategy treats customers similarly despite significant differences in purchasing behavior and customer value. Without a structured understanding of customer segments, marketing campaigns become less effective, customer engagement decreases, and opportunities for revenue growth remain untapped.

# 6. Feature Categorization (Data Dictionary)

To better understand the dataset, the available features are grouped into business-oriented categories rather than being interpreted individually.

This categorization improves readability and provides a clear understanding of the information available for customer analysis, feature engineering, and customer segmentation in the later stages of the project.

The dataset consists of the following feature groups:

- Customer Demographics
- Customer Relationship
- Product Spending
- Purchase Behaviour
- Marketing Campaign Interaction
- Administrative Variables





## 6.1 Customer Demographics

These features describe the customer's personal and socio-economic characteristics.

| Feature | Description |
|----------|-------------|
| **ID** | Unique customer identifier |
| **Year_Birth** | Customer's birth year |
| **Education** | Highest education level |
| **Marital_Status** | Customer's marital status |
| **Income** | Annual household income |
| **Kidhome** | Number of children living at home |
| **Teenhome** | Number of teenagers living at home |

## 6.2 Customer Relationship

These features describe the customer's relationship with the company.

| Feature | Description |
|----------|-------------|
| **Dt_Customer** | Date the customer enrolled with the company |
| **Recency** | Number of days since the customer's most recent purchase |

## 6.3 Product Spending

These features represent the amount spent by each customer across different product categories.

| Feature | Description |
|----------|-------------|
| **MntWines** | Amount spent on wine products |
| **MntFruits** | Amount spent on fruits |
| **MntMeatProducts** | Amount spent on meat products |
| **MntFishProducts** | Amount spent on fish products |
| **MntSweetProducts** | Amount spent on sweet products |
| **MntGoldProds** | Amount spent on gold products |

## 6.4 Purchase Behaviour

These features capture how customers interact with different purchasing channels.

| Feature | Description |
|----------|-------------|
| **NumDealsPurchases** | Purchases made using discounts or special deals |
| **NumWebPurchases** | Purchases made through the website |
| **NumCatalogPurchases** | Purchases made through the product catalog |
| **NumStorePurchases** | Purchases made in physical stores |
| **NumWebVisitsMonth** | Number of website visits in the last month |

## 6.5 Marketing Campaign Interaction

These variables measure customer engagement with previous marketing campaigns.

| Feature | Description |
|----------|-------------|
| **AcceptedCmp1** | Accepted Campaign 1 |
| **AcceptedCmp2** | Accepted Campaign 2 |
| **AcceptedCmp3** | Accepted Campaign 3 |
| **AcceptedCmp4** | Accepted Campaign 4 |
| **AcceptedCmp5** | Accepted Campaign 5 |
| **Complain** | Indicates whether the customer has submitted a complaint |
| **Response** | Indicates whether the customer accepted the most recent campaign |

## 6.6 Administrative Variables

These variables appear to be administrative fields originating from the company's internal database.

During the initial inspection, both variables were found to contain only a single unique value across all customer records. Since they exhibit zero variance, they are unlikely to contribute meaningful information for customer segmentation or downstream analysis and will be evaluated further during the data preprocessing stage.

| Feature | Description |
|----------|-------------|
| **Z_CostContact** | Administrative constant field (requires validation) |
| **Z_Revenue** | Administrative constant field (requires validation) |
